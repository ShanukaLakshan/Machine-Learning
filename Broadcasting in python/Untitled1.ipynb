## Activation functions



## Sigmoid function

```python
import numpy as np

# sigmid activation function
def sigmoid(x):
    return ( 1.0 / (1.0 + np.exp(-x)))

inputs = [13,3,-12,0]

for x in inputs:
    print("{}\t-->\t{}".format(x,sigmoid(x)))
```

## Tanh function

```python
import numpy as np

# tanh activation function

def tanh(x):
    return ( (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))) 

inputs = [13,3,-12,0]

for x in inputs:
    print("{}\t-->\t{}".format(x,tanh(x)))
```

## ReLU function

```python
import numpy as np

# relu activation function

def relu(x):
    return ( max(0.0,x)) 

inputs = [13,3,-12,0]

for x in inputs:
    print("{}\t-->\t{}".format(x,relu(x)))
```

## Leaky ReLU function

```python
import numpy as np

# leaky relu activation function

def leakyRelu(x):
    return ( max(0.1*x,x)) 

inputs = [13,3,-12,0]

for x in inputs:
    print("{}\t-->\t{}".format(x,leakyRelu(x)))
```

## Softmax function


```python
import numpy as np

# softmax activation function

def softmax(x):
    return (np.exp(x) / np.exp(x).sum())  

inputs = [13,3,-12,0]

print(softmax(inputs))

for x in inputs:
    print("{}\t-->\t{}".format(x,softmax(x)))
```
