## Ordinal Encoding

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/siddiquiamir/Python-Data-Preprocessing/main/ordinal.csv")
```

```python
df
```

```python
df["Education"].unique()
```

```python
enc = OrdinalEncoder()
```

```python
enc.fit_transform(df[["Education"]])
```

```python
df.head()
```

```python
education = ['PhD', 'Masters', 'Bachelors']
```

```python
enc = OrdinalEncoder(categories=[education])
```

```python
enc.fit_transform(df[["Education"]])
```

```python
df[["Education"]] = enc.fit_transform(df[["Education"]])
```
